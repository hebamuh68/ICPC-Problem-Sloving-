```python
def difString(str):
   count = 0
   for i in range(len(str) - 2):
       for j in range(i + 1, len(str) - 1):
           a = str[0:i + 1]
           b = str[i + 1: j + 1]
           c = str[j + 1: len(str)]

           if (a + b != b + c) and (b + c != c + a) and (a + b != c + a):
               count += 1

   return count
```

```python
def sawtooth(arr):
   if len(set(arr)) == 1:
       return 0

   if (arr == sorted(arr)) or (arr[::-1] == sorted(arr)):
       return len(arr) - 1

   count = 0
   for i in range(len(arr) - 2):
       if ((arr[i] > arr[i + 1]) and (arr[i + 1] < arr[i + 2])) or (
               (arr[i] < arr[i + 1]) and (arr[i + 1] > arr[i + 2])):
           n = i + 2
           count += ((n * (n + 1)) // 2) - n

   return count
```

<b>strictly ascending</b>
- Each element is strictly larger (not just merely equal to) than the element that precedes i.
- Empty sequence is considered ascending, as is also every one-element sequence.

<b>Examples </b>


```python
def strictly_ascending(nums):
    
    if len(nums) <= 1:
        return True
    
    start = 1
    end = -1
    temp = nums[0]
    
    for i in range(1, len(nums)):
        if i%2 == 0:
            if nums[start] > temp:
                temp = nums[start]
                start += 1
            else:
                return False
            
        else:
            if nums[end] > temp:
                temp = nums[end]
                end -= 1
            else:
                return False
    return True

strictly_ascending([1,3,4,5,6])
```

```python
def solution(queryType, query):
    hashmap = {}
    
    for i in range(len(queryType)):
        
        if queryType[i] == "insert":
            key, val = query[i][0], query[i][1]
            hashmap[key] = val
        
        elif queryType[i] == "addToValue":
            for key in hashmap.keys():
                hashmap[key] += query[i][0]
        
        elif queryType[i] == "addToKey":
            pass
                        
        elif queryType[i] == "get":
                count = 0
                for key in hashmap.keys():
                    count += hashmap[key]
                
                return count

    

            
queryType = ["insert", "insert", "addToValue", "addToKey", "get"]
query = [[1,2],[2,3], [2], [1], [3]]

print(solution(queryType, query))
```

```python
def mergeStrings(s1, s2):
    
    # record appear times
    record1 = {}
    record2 = {}
    for ch in s1:
        record1[ch] = record1.get(ch,0)+1
    for ch in s2:
        record2[ch] = record2.get(ch,0)+1
    
    # merge with two pointers
    pt1 = 0
    pt2 = 0
    len1 = len(s1)
    len2 = len(s2)
    res = []
    while pt1 < len1 and pt2 < len2:
        if record1[s1[pt1]] < record2[s2[pt2]]:
            res.append(s1[pt1])
            pt1 += 1
        elif record1[s1[pt1]] > record2[s2[pt2]]:
            res.append(s2[pt2])
            pt2 += 1
        else:
            # if equal times
            if s1[pt1] <= s2[pt2]:
                res.append(s1[pt1])
                pt1 += 1
            elif s1[pt1] > s2[pt2]:
                res.append(s2[pt2])
                pt2 += 1
                
    # continue with the remaining characters
    while pt1 < len1:
        res.append(s1[pt1])
        pt1 += 1
    
    while pt2 < len2:
        res.append(s2[pt2])
        pt2 += 1
    return "".join(res)
 
s = "dce"
ch = 'cccbd'
print(mergeStrings(s, ch))
```
